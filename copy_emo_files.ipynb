```python
import json
from pathlib import Path
import random
from shutil import copy
```

```python
with open("emo_speaker_file.json") as f:
    emo_dict = json.load(f)
```

```python
data_path = Path("data/esd/resampled")
```

```python
!rm -rf data/origin
```

```python
save_path = Path("data/origin")
for emo, speaker_filelist in emo_dict.items():
    for speaker, files in speaker_filelist.items():
        if speaker == "0019":
            continue
        for file in random.choices(files, k=7):
            open_path = data_path / speaker / f"{file}.wav"
            save_wav = save_path / speaker / emo 
            save_wav.mkdir(exist_ok=True, parents=True)
            copy(open_path, save_wav / f"{file}.wav")
```
